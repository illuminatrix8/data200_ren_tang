**1c.** Present a time-series line chart with the following elements (10 points)  

- The chart needs to have two lines - one for Bernie Sanders and one for Hillary Clinton.
- Show the date on the x-axis
- Add vertical lines for these dates of the primary debates: 17 Jan 2016, 11 Feb 2016, 9 Mar 2016
- Show the total contribution amount for that day on the y-axis
- Include a title
- Include axis labels

How did the debates effect fundraising / donations? Did fundraising increase or decrease after the debates for each candidate?

```python
# 1c YOUR CODE HERE
def plot_time_series():
    # Load the dataset
    df = pd.read_csv('../2016_ca_primary_cleaned.csv', index_col=False, parse_dates=['contb_receipt_dt'])
    
    # Filter data for Bernie Sanders and Hillary Clinton
    bernie_data = df[df['cand_nm'] == 'Sanders, Bernard']
    hillary_data = df[df['cand_nm'] == 'Clinton, Hillary Rodham']
    
    # Group data by date and calculate total contribution amount for each candidate
    bernie_daily_total = bernie_data.groupby('contb_receipt_dt')['contb_receipt_amt'].sum()
    hillary_daily_total = hillary_data.groupby('contb_receipt_dt')['contb_receipt_amt'].sum()
    
    # Create the time-series line chart
    plt.figure(figsize=(12, 6))
    plt.plot(bernie_daily_total.index, bernie_daily_total.values, label='Bernie Sanders')
    plt.plot(hillary_daily_total.index, hillary_daily_total.values, label='Hillary Clinton')
    
    # Add vertical lines for debate dates
    debate_dates = ['2016-01-17', '2016-02-11', '2016-03-09']
    for date in debate_dates:
        plt.axvline(pd.to_datetime(date), color='gray', linestyle='--', label=f'Debate Date: {date}')
    
    # Set labels and title
    plt.xlabel('Date')
    plt.ylabel('Total Contribution Amount')
    plt.title('Time-Series of Total Contribution Amount by Candidate')
    
    # Show legend
    plt.legend()
    
    # Show the plot
    plt.show()

# Call the function to generate the time-series chart
plot_time_series()
```

- **1c answer here:** 

# The debates are correlated with a relative increase in contribution amount for Bernie Sanders, compared with Hilary Clinton.

```python
# 2d YOUR CODE HERE
def show_both(df):
    # Load the dataset
    df = pd.read_csv('../2016_ca_primary_cleaned.csv', index_col=False, parse_dates=['contb_receipt_dt'])

    # Group the data by candidate and count the number of donations
    donation_counts = df.groupby('cand_nm')['contb_receipt_amt'].count().reset_index()
    donation_counts.columns = ['Candidate', 'Number of Donations']

    # Group the data by candidate and calculate the total value of contributions
    total_contributions = df.groupby('cand_nm')['contb_receipt_amt'].sum().reset_index()
    total_contributions.columns = ['Candidate', 'Total Contribution Amount']

    # Merge the two DataFrames on the 'Candidate' column
    combined_table = pd.merge(donation_counts, total_contributions, on='Candidate')

    combined_table['Average Donation'] = combined_table['Total Contribution Amount'] / combined_table['Number of Donations']

    # Sort the combined table by the number of donations in descending order
    combined_table = combined_table.sort_values(by='Average Donation', ascending=False)


    # Display the combined table
    print(combined_table)

show_both(contrib)
```

**2e. Plotting a Bar Chart** (10 points)

Make a single bar chart that shows two different bars per candidate with one bar as the total value of the donations and the other as average $ per donation. 
- Show the Candidates Name on the x-axis
- Show the amount on the y-axis
- Include a title
- Include axis labels
- Sort by total value of the donations
- Make the y-axis a log-scale to show both numbers! (matplotlib docs: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.yscale.html )

```python
# 2e YOUR CODE HERE
def compare_donation_total_and_avg():
    # Load the dataset
    df = pd.read_csv('../2016_ca_primary_cleaned.csv', index_col=False, parse_dates=['contb_receipt_dt'])

    # Group the data by candidate and calculate the total value of contributions and average donation
    candidate_data = df.groupby('cand_nm').agg({
        'contb_receipt_amt': ['sum', 'count']
    }).reset_index()
    candidate_data.columns = ['Candidate', 'Total Contribution Amount', 'Number of Donations']

    # Calculate the average donation amount per candidate
    candidate_data['Average Donation'] = candidate_data['Total Contribution Amount'] / candidate_data['Number of Donations']

    # Sort the data by total value of contributions
    candidate_data = candidate_data.sort_values(by='Total Contribution Amount', ascending=False)

    # Create the bar chart
    plt.figure(figsize=(12, 6))
    plt.bar(candidate_data['Candidate'], candidate_data['Total Contribution Amount'], label='Total Contribution Amount', log=True)
    plt.bar(candidate_data['Candidate'], candidate_data['Average Donation'], label='Average $ per Donation', log=True, alpha=0.7)

    # Set labels and title
    plt.xlabel('Candidates')
    plt.ylabel('Amount (log scale)')
    plt.title('Total Contribution Amount vs. Average $ per Donation by Candidate')

    # Show axis labels on the x-axis
    plt.xticks(rotation=45, ha='right')

    # Show legend
    plt.legend()

    # Show the plot
    plt.tight_layout()
    plt.show()

compare_donation_total_and_avg()
```

**3c.** Now run the `get_donors` function from 3b on subsets of the dataframe corresponding to three candidates. Show each of the three candidates below. (5 points)

```python
# 3c YOUR CODE HERE
def compare_three():
    # Load the dataset
    df = pd.read_csv('../2016_ca_primary_cleaned.csv', index_col=False, parse_dates=['contb_receipt_dt'])

    # Define the function to get top 5 occupations
    def get_donors(dataframe):
        top_occupations = dataframe['contbr_occupation'].value_counts().head(5)
        return top_occupations

    # Subset the data for Hillary Clinton, Bernie Sanders, and Donald Trump
    hillary_data = df[df['cand_nm'] == 'Clinton, Hillary Rodham']
    bernie_data = df[df['cand_nm'] == 'Sanders, Bernard']
    trump_data = df[df['cand_nm'] == 'Trump, Donald J.']

    # Get the top 5 occupations for each candidate using the function
    top_occupations_hillary = get_donors(hillary_data)
    top_occupations_bernie = get_donors(bernie_data)
    top_occupations_trump = get_donors(trump_data)

    # Display the top 5 occupations for each candidate
    print("Top 5 Occupations for Hillary Clinton:")
    print(top_occupations_hillary)

    print("\nTop 5 Occupations for Bernie Sanders:")
    print(top_occupations_bernie)

    print("\nTop 5 Occupations for Donald Trump:")
    print(top_occupations_trump)

compare_three()
```

**3d.** Finally, use `groupby` to separate the entire dataset by candidate. (10 points)

- Call .apply(get_donors) on your groupby object, which will apply the function you wrote to each subset of your data.
- Look at your output and marvel at what pandas can do in just one line!

```python
# 3d YOUR CODE HERE

def analyze_top_donor_occupation():

    # Load the dataset
    df = pd.read_csv('../2016_ca_primary_cleaned.csv', index_col=False, parse_dates=['contb_receipt_dt'])

    # Define the function to get top 5 occupations
    def get_donors(dataframe):
        top_occupations = dataframe['contbr_occupation'].value_counts().head(5)
        return top_occupations

    # Group the entire dataset by candidate and apply the function to each subset
    top_occupations_by_candidate = df.groupby('cand_nm').apply(get_donors)

    # Display the top 5 occupations for each candidate
    print(top_occupations_by_candidate)

analyze_top_donor_occupation()
```
