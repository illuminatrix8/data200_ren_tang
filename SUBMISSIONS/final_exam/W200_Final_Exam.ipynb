### VCF file description (Summarized from version 4.1)

```
* The VCF specification:

VCF is a text file format which contains meta-information lines, a header line, and then data lines each containing information about a position in the genome. The format also can contain genotype information on samples for each position.

* Fixed fields:

There are 8 fixed fields per record. All data lines are **tab-delimited**. In all cases, missing values are specified with a dot (.). 

1. CHROM - chromosome number
2. POS - position DNA nuceleotide count (bases) along the chromosome
3. ID - The unique identifier for each mutation
4. REF - reference base(s)
5. ALT - alternate base(s)
6. FILTER - filter status
7. QUAL - quality
8. INFO - a semicolon-separated series of keys with values in the format: <key>=<data>

```
### Applicable INFO field specifications

```
GENEINFO = <Gene name>
CLNSIG =  <Clinical significance>
CLNDN = <Disease name>
```

````python
# Part 1: Remove Metadata

def read_data(file_path):
    """Read the file and return lines without metadata."""
    with open(file_path, 'r') as file:
        return [line.strip() for line in file if not line.startswith('#')]

# Path to the ClinVar file
file_path = '../../clinvar_final.txt'

# Read data from the file
lines = read_data(file_path)

# Verification: Print the first 5 lines to verify
print("First 5 lines of data (without metadata):")
for line in lines[:5]:
    print(line)
````

````python
# Part 2: Separate Columns in Text into DataFrame Columns

import pandas as pd

def create_initial_dataframe(lines):
    """Create a DataFrame from the fixed fields."""
    columns = ['Chromosome', 'Position', 'Mutation ID', 'Reference Base', 'Alternate Base', 'Filter', 'Quality', 'Info']
    data = [line.split('\t') for line in lines]
    return pd.DataFrame(data, columns=columns)

# Create initial DataFrame
initial_df = create_initial_dataframe(lines)

# Verification: Display the first few rows of the DataFrame
print("Initial DataFrame with separated columns:")
print(initial_df.head())
````

````python
# Part 3: Unpack the INFO Column - v2 - Vectorized Approach

def extract_info_keys(df):
    """Extract all unique keys from the 'Info' column."""
    all_keys = set()
    for info_str in df['Info']:
        keys = [pair.split('=')[0] for pair in info_str.split(';') if '=' in pair]
        all_keys.update(keys)
    return all_keys

def parse_info_field_vectorized(info_str, key):
    """Vectorized parsing of the INFO field for a given key."""
    return info_str.str.extract(f'{key}=([^;]*)')[0].fillna('Not_Given')

def unpack_info_column_vectorized(df, info_keys):
    """Vectorized expansion of the Info column into separate columns."""
    for key in info_keys:
        df[key] = parse_info_field_vectorized(df['Info'], key)
    df.drop('Info', axis=1, inplace=True)
    return df

# Identify all unique keys in the 'Info' column
info_keys = extract_info_keys(initial_df)

# Unpack the Info column using a vectorized approach
clinvar_df = unpack_info_column_vectorized(initial_df, info_keys)

# Verification: Display the first few rows of the DataFrame
print("DataFrame with unpacked INFO column (Vectorized):")
print(clinvar_df.head())
````

````python
# Function to extract gene name from GENEINFO
def parse_gene_name(geneinfo):
    return geneinfo.split(':')[0] if ':' in geneinfo else 'Not_Given'

# Function to analyze harmful mutations
def analyze_harmful_mutations(df, harmful_criteria):
    # Extract gene name
    df['Gene Name'] = df['GENEINFO'].apply(parse_gene_name)

    # Filter DataFrame for harmful mutations based on CLNSIG
    harmful_mutations = df[df['CLNSIG'].isin(harmful_criteria)].copy()

    # Replace missing values with 'Not_Given'
    harmful_mutations.fillna('Not_Given', inplace=True)

    # Limit to the first 100 harmful mutations
    top_100_harmful = harmful_mutations.head(100)

    # Make sure that you only give your boss the dangerous mutations and include:
    # 1) Gene name
    # 2) Mutation ID number
    # 3) Mutation Position (chromosome & position)
    # 4) Mutation value (reference & alternate bases)
    # 5) Clinical significance (CLNSIG)
    # 6) Disease that is implicated

    # Filter to required columns
    top_100_harmful = top_100_harmful[['Gene Name', 'Mutation ID', 'Chromosome', 'Position', 'Reference Base', 'Alternate Base', 'CLNSIG', 'CLNDN']]

    # Count total harmful mutations
    total_harmful_count = len(harmful_mutations)

    # Displaying the results
    print(f"\nTotal Count of Harmful Mutations: {total_harmful_count}")
    print("First 100 Harmful Mutations:")
    return top_100_harmful

# Assumed harmful mutation criteria
harmful_criteria = ['Pathogenic', 'Likely_pathogenic']  # Add more terms as needed


# Call the function with the DataFrame and criteria
df = analyze_harmful_mutations(clinvar_df, harmful_criteria)

df
````
